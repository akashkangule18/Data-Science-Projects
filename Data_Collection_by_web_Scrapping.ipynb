```python
# importing Important Libraries
import pandas as pd
import numpy as np
import time
import re
from bs4 import BeautifulSoup
import requests
```

```python
# loading data form web in text format by using Beautiful Soup
final_df = pd.DataFrame()

for i in range(1,100):

  url = "https://www.flipkart.com/search?q=smartphones&page={}".format(i)
  headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0 Safari/537.36"
    }
  response = requests.get(url,headers=headers)

  # Getting HTML text from web by creating a soup object
  soup = BeautifulSoup(response.text,'lxml')

  # Fetching Data
  all_divs = soup.find_all('div',class_='tUxRFH')

  Phone_name = []
  Discription = []
  Stars = []
  Total_reviews = []
  Price = []



  for phone in all_divs:

  # for phone name:
    try:
      Phone_name.append(phone.find('div',class_='KzDlHZ').text.strip())
    except:
      Phone_name.append(np.nan)


  # fetching Discription
    dis = phone.find('ul',class_ = 'G4BRas')
    Discription.append(dis.text.strip() if dis else None)

  # Getting Stars
    try:
      Stars.append(phone.find('div',class_='XQDdHH').text.strip())
    except:
      Stars.append(np.nan)


  # Getting Reviews and ratings
    try:
      Total_reviews.append(phone.find('span',class_='Wphh3N').find('span').text.strip())
    except:
      Total_reviews.append(np.nan)

  # fetching Price
    pr = phone.find('div',class_='Nx9bqj _4b5DiR')
    Price.append(pr.text.strip() if pr else None)

  df = pd.DataFrame({
        'phone name' : Phone_name,
        'discription' : Discription,
        'stars' : Stars,
        'total_reviews' : Total_reviews,
        'price':Price
    })

  final_df = pd.concat([final_df,df],ignore_index=True)


# Preview Of uncleaned data
final_df.head()

```

```python
# Removing Duplicted value
final_df.drop_duplicates(inplace=True)

# Converting Final_df into csv format.
final_df.to_csv('Uncleaned SmartPhones.csv')
```

# ***Cleaning Dataset***




```python
# Loading and making copy of data
data = pd.read_csv("/content/Uncleaned SmartPhones.csv")
df = data.copy()
df.head()
```

```python
# Adding an indexing column to ensure observation of data in term of rows is correct from spreadsheet.
df = df.reset_index()
df['index']  = df['index'] + 2
```

```python
# removing unwanted column
df.drop(columns=['Unnamed: 0'],inplace=True)
df.head()
```

```python
df.info()
```

```python
# Removing Unwanted symbols from price column
df['price'] = df['price'].str.replace('₹','').str.replace(',','')
df.head()
```

```python
# there are missing values in stars, total_reviews and price column. So Right now we going to fill nan values with 0.
df['stars'] = df['stars'].fillna(0)
df['total_reviews'] = df['total_reviews'].fillna(0)
df['price'] = df['price'].fillna(0)
df.info()
```

```python
# Converting price column into int
df['price'] = df['price'].astype('int')

# Sucessfully Converted
df.info()
```

```python
# Now as per my observation in this data the most problematic rows are from 243-266 and 255,382,386.
# In these rows the info is not about smartphones. it contains the info. batteries and feature phones. So need to remove them
# again according to price values the price of batteries is less than 380 and price for feaature phones is less than 3500.
# So now i am going to sort the data on the basis of these price range.

df = df[df['price'] >= 3500]

# Again Creaating copy -----> Not necessary at this stage. Just for Self Study
new_data = df.copy()
new_data.head()
```

```python
# Extracting Color Column From Phone name seperately
new_data['phone_color'] = new_data['phone name'].str.extract(r"\(([A-Z][a-z]+ [A-Z][a-z]+),")
new_data.head()
```

```python
# Applying Split Function on the phone name coulumn
new_data['phone name'] = new_data['phone name'].str.split('(').str.get(0)
new_data.head()
```

```python
# Now we need to add more columns by fetching information from the discription column
new_data['RAM']=new_data['discription'].str.split("|").str.get(0).str.split(" ").str.get(0)
new_data['ROM']=new_data['discription'].str.split("|").str.get(1).str.split(" ").str.get(1)
new_data.head(5)
```

```python
# Phone Size
new_data['phone_size_in_inch'] = new_data['discription'].str.findall(r"\((\d+\.\d+ [a-z]+)\)").str.get(0).str.split(" ").str.get(0)
new_data.head()
```

```python
# Front Camera Details
new_data['Front Camera in MP']=new_data['discription'].str.findall(r"(\d+MP)").str.get(0).str.findall(r"(\d+)").str.get(0)
new_data.head()
```

```python
# Separeting Total ratings and Total reviews from discription column.
new_data['ratings']=new_data['total_reviews'].str.split("&").str.get(0).str.split(" ").str.get(0).str.replace(",","")
new_data['reviews']=new_data['total_reviews'].str.split("&").str.get(1).str.split(" ").str.get(0).str.replace(",","")
```

```python
# dropping unwanted column
new_data.drop(columns=['index','total_reviews','discription'],inplace=True)
```

```python
# Changing Data types of columnns

new_data['RAM'] = pd.to_numeric(new_data['RAM'], errors='coerce').fillna(0).astype('int')
new_data['ROM'] = pd.to_numeric(new_data['ROM'], errors='coerce').fillna(0).astype('int')
new_data['phone_size_in_inch'] = pd.to_numeric(new_data['phone_size_in_inch'], errors='coerce').fillna(0).astype('float')
new_data['Front Camera in MP'] = pd.to_numeric(new_data['Front Camera in MP'], errors='coerce').fillna(0).astype('int')
new_data['ratings'] = pd.to_numeric(new_data['ratings'], errors='coerce').fillna(0).astype('int')
new_data['reviews'] = new_data['reviews'].str.replace('\xa0', '').str.replace(',', '')
new_data['reviews'] = pd.to_numeric(new_data['reviews'], errors='coerce').fillna(0).astype('int')

```

```python
# Sucessfully Changed Data types
new_data.info()
```

```python
# Convering Clean Data into csv format
new_data.to_csv('Cleaned Smartphones.csv')
```
